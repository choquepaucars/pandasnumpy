# Agregar o eliminar datos

**Borrar fila y columnas**
- El metodo head(n=2): muestra la cabeza de dataframe en este caso indicamos que muestre 2 elementos

- El metodo drop(labels, axis): Elimina una fila o columna indicada en axis y para eliminar se espeficica la columna a eliminar con el parametro labels, y con el parametro inplace=True: Hacemos que se elimine definitivamente del array

- del de python: del df_books["Price"]: Este comando nos ayuda a eliminar la columna Price
- df_books.drop(labels=0, axis=0).head(2): borra la primera fila del array

**Agregar filas y columnas**
Podemos agregar columnas de la siguiente forma:
```python
import numpy as np
import pandas as pd
df_books["nuevacolum"]=np.nan
```

Podemos agregar filas de la siguiente manera:
```python
df_books.append(df_books)
```
```python
df_books.loc[551,:]=np.nan
```
