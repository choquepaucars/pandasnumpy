## concat

```python
df1=pd.DataFrame(data={
```

## merge
El metodo merge(): une dataframes segun las espeficicaciones en los parametros
merge del df llamado izq y der; es muy importante poner tambien que una tomando en cuenta una columna(en este caso key)
```python
izq.merge(right=der, on="key")

izq.merge(right=der, left_on="Key", right_on="Key_2")


# En el parametro how tenemos que especificar como va traer: left, right, ineer, outer
izq.merge(right=der, left_on="key", right_on="key_2", how="left") 
```

## join


join realiza el index math. busca a travez de indices

```python

izq = pd.DataFrame(data={
    "A": [f"A{i}" for i in np.arange(0, 3) if i == i],
    "B": [f"B{i}" for i in np.arange(0, 3) if i == i]},
    index=["k0", "k1", "k2"]
)

der = pd.DataFrame(data={
    "C": [f"C{i}" for i in np.arange(0, 3) if i == i],
    "D": [f"D{i}" for i in np.arange(0, 3) if i == i]},
    index=["k0", "k2", "k3"]
)

 # y puedes especificar como vas a unir con join
izq.join(other=der, how="left")
->

    A	B	C	D
k0	A0	B0	C0	D0
k1	A1	B1	NaN	NaN
k2	A2	B2	C1	D1
```

````python
izq = pd.DataFrame(data={
    "key": ["k0", "k1", "k2", "k3"],
    "A": [f"A{i}" for i in np.arange(0, 4) if i == i],
    "B": [f"B{i}" for i in np.arange(0, 4) if i == i]
})

der = pd.DataFrame(data={
    "key_2": ["k0", "k1", "k2", "k3"],
    "C": [f"C{i}" for i in np.arange(0, 4) if i == i],
    "D": [f"D{i}" for i in np.arange(0, 4) if i == i]
})
````

````python
izq = pd.DataFrame(data={
    "key": ["k0", "k1", "k2", np.nan],
    "A": [f"A{i}" for i in np.arange(0, 4) if i == i],
    "B": [f"B{i}" for i in np.arange(0, 4) if i == i]
})

der = pd.DataFrame(data={
    "key_2": ["k0", "k1", "k2", "k3"],
    "C": [f"C{i}" for i in np.arange(0, 4) if i == i],
    "D": [f"D{i}" for i in np.arange(0, 4) if i == i]
})


izq.merge(right=der, left_on="key", right_on="key_2", how="outer")
````
