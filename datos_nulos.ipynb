# Valores nulos
- El metodo isnull(): devuelve valores bool según si el datos es NaN(True) y nosenan(False)
- df.isnull()*1: devuelve valores buleanos en 0 y 1(True)
- El metodo fillna(): Rellena los valores nan con otro valor
- df.fillna(value=df.mean()): los valores nan a menudo se rellena con la media
- df.interpolate(): con el metodo interpolate esta función asume que se sigue una serie y recomiendo un dato
- df.dropna(): muchos cientificos terminana eliminando los valores nan analizando si son relevantes o no

```python
import pandas as pd
import numpy as np
```

```python
dict={"col1":[1,2,3,np.nan],
"col2": [4,np.nan,6,7],
"calo3":["a","b","c",None]}

df=pd.DataFrame(data=dict)
df
```

```python
df.isnull()
```

```python
df.isnull()*1
```

```python
df.fillna(value="Missing")
```

```python
df.fillna(value=df.mean())
```

```python
df.interpolate()
```

```python
df.dropna()
```

# Filtrado por condiciones

```python
import pandas as pd
```

```python
df_books=pd.read_csv(filepath_or_buffer="./data/bestsellers_data.csv", sep=",", header=0)
df_books.head(5)
```

```python
mayor_a_2016=df_books["Year"]>2016
```

```python
df_books[df_books["Year"]>2016]
```

```python
genre_fiction=df_books["Genre"]=="Fiction"
```

```python
df_books[genre_fiction & mayor_a_2016]
```

```python
df_books[df_books.Year.eq(2016)]
```
